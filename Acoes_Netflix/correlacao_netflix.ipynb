##### Correlação

```python
features = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]
```

```python
corr_matrix = df[features].corr()
corr_matrix
```

```python
plt.figure(figsize=(18, 10))
sns.heatmap(corr_matrix, annot=True, linewidth=.5, fmt=".2f")
plt.show()
```
